# DATA EXPLORATION

```python
import pandas as pd

# Load the fetched movie data
movie_data = pd.read_csv('/content/fetched_movie_data.csv')

# Explore the dataset
print(movie_data.head())
print(movie_data.info())
print(movie_data.describe())
```

```python
# Analyze the distribution of movie genres
genre_counts = movie_data.explode('genres')['genres'].value_counts()
print("Genre Distribution:")
print(genre_counts)
```

```python
# Visualize the genre distribution
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
genre_counts.plot(kind='bar')
plt.title('Distribution of Movie Genres')
plt.xlabel('Genre')
plt.ylabel('Number of Movies')
plt.show()
```

```python
# Analyze the distribution of ratings
plt.figure(figsize=(10, 6))
movie_data['rating'].plot(kind='hist', bins=20, edgecolor='black')
plt.title('Distribution of Movie Ratings')
plt.xlabel('Rating')
plt.ylabel('Frequency')
plt.show()
```
